````python
from collections import Counter

cols = ['BRANDS', 'SUPERGROUP', 'RPM Name', 'eRPM Name', 'ACUTE_CHRONIC']
df_req = df_ims[cols]

di_req = Counter(df_req['BRANDS'].values)
print(di_req)
````

````python
# Taking only unique brand values

unique_brands = []

for brand, count in di_req.items():
    if count == 1:
        unique_brands.append(brand)
````

````python
print("Total number of unique brands: ", len(unique_brands))
````

````python
from pprint import pprint

pprint(unique_brands)
````

````python
df_final = df_req[df_req['BRANDS'].isin(unique_brands) == True]
````

````python
df_final[:40]
````

````python
print("Found duplicated values: ", round((len(df_final)/len(df_req))*100, 2), "%")
````

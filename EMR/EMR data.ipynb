```python
lab=lab[lab['new_name'].isin(lab_value_count.index) == True]
```

```python
# Names have the same meaning

lab.loc[lab['new_name'].str.contains('|'.join(['post','pp']))==True,"new_name"]= "ppg"
lab.loc[lab['new_name'].str.contains('|'.join(['fasti','fbs']))==True,"new_name"]= "fbs"
```

```python
lab.loc[lab['new_name'].str.contains('|'.join(['diastolic','diastolic']))==True,"new_name"]= "bp_diastolic"
lab.loc[lab['new_name'].str.contains('|'.join(['systolic','systolic']))==True,"new_name"]= "bp_systolic"
```

```python
lab.loc[lab['new_name'].str.contains('|'.join(['ldl','ldl']))==True,"new_name"]= "ldl"
lab.loc[lab['new_name'].str.contains('|'.join(['hdl','hdl']))==True,"new_name"]= "hdl"
lab.loc[lab['new_name'].str.contains('|'.join(['tri','tri']))==True,"new_name"]= "triglycerides"
```
